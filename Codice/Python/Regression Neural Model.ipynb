```python
import os.path 
import numpy as np
import pandas as pd
import csv

from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error,r2_score,mean_squared_error
import tensorflow as tf

from tensorflow import keras
from tensorflow.keras import layers
```

```python
Boston=pd.read_csv(r'D:\Users\aumar\Documents\Estudios Milano\PACS\Proyecto\Codice\Data\Boston.csv',header=0,names=["index","crim","zn","indus","chas","nox","rm","age","dis","rad","tax","ptratio","black","lstat","medv"])
Boston=Boston.drop(["index"],axis=1)
Boston.head()
```

```python
train_dataset = Boston.sample(frac=0.6,random_state=0)
test_dataset = Boston.drop(train_dataset.index)
val_dataset=test_dataset.sample(frac=0.5,random_state=0)
test_dataset=test_dataset.drop(val_dataset.index)
```

```python
train_labels = train_dataset.pop('medv')
test_labels = test_dataset.pop('medv')
val_labels = val_dataset.pop('medv')
```

```python
 scaler = StandardScaler()
scaler.fit(train_dataset)

scaled_train=scaler.transform(train_dataset)
scaled_test=scaler.transform(test_dataset)
scaled_val=scaler.transform(val_dataset)
```

# Model Definition

```python
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(64, activation='relu', input_dim=len(train_dataset.columns)))
model.add(tf.keras.layers.Dense(64, activation='relu'))
model.add(tf.keras.layers.Dense(1))
```

```python
opt=tf.keras.optimizers.SGD(0.001, momentum=0.9)
```

```python
model.compile(optimizer=opt, loss='mse')
```

```python
EVALUATION_INTERVAL = 3000
EPOCHS = 1000
```

```python
History=model.fit(scaled_train, train_labels,
  epochs=EPOCHS, validation_data=(scaled_val,val_labels))
```

# Test Analysis

```python
train_predictions=model.predict(scaled_train).flatten()
test_predictions = model.predict(scaled_test).flatten()
```

```python
mse = mean_squared_error(test_labels, test_predictions)
print('Mean Squared Error: ',mse)
mae = mean_absolute_error(test_labels, test_predictions)
print('Mean Absolute Error: ',mae)
rsq = r2_score(train_labels,train_predictions) #R-Squared on the training data
print('R-square, Training: ',rsq)
rsq = r2_score(test_labels,test_predictions) #R-Squared on the testing data
print('R-square, Testing: ',rsq)
```
